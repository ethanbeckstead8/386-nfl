```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

```python
base_url = "https://www.pro-football-reference.com"
team_url1 = "https://www.pro-football-reference.com/teams/"
team = "kan"
team_url2="/"
team_url3="/gamelog/"

kansasBox = []
for year in range(2019, 2024):
    time.sleep(2)
    url = team_url1 + team + team_url2 + str(year) + team_url3
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for cell in soup.find_all('td', {'data-stat': 'boxscore_word'}):
        link_tag = cell.find('a')
        if link_tag:
            kansasBox.append(base_url + link_tag['href'])

print("Kansas Boxscore Links:", kansasBox)
team = "min"
minnesotaBox = []
for year in range(2019, 2024):
    time.sleep(2)
    url = team_url1 + team + team_url2 + str(year) + team_url3
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for cell in soup.find_all('td', {'data-stat': 'boxscore_word'}):
        link_tag = cell.find('a')
        if link_tag:
            # Append the full URL for the box score page
            minnesotaBox.append(base_url + link_tag['href'])
print("Minnesota Boxscore Links:", minnesotaBox)

```

```python
# Depending on what stats we want
def getTeamStatsTable(url):
    time.sleep(2)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    tables = soup.find_all('table')
    game_stats = {'URL': url}
    player_stats = []
    if len(tables)<3:
        return "none"
    table = tables[2]
    if table:
        headers = [th.get_text() for th in table.find('thead').find_all('th')]
        for row in table.find('tbody').find_all('tr'):
            player_data = {}
            cells = row.find_all(['th', 'td'])
            for header, cell in zip(headers, cells):
                player_data[header] = cell.get_text(strip=True)
            player_stats.append(player_data)
        return player_stats
    return "none"
        
        
```

```python
offenseStatsKan = []
for link in kansasBox:
    offenseStatsKan.append(getTeamStatsTable(link))
    print(link)
```

```python
offenseStatsMin = []
for link in minnesotaBox:
    offenseStatsMin.append(getTeamStatsTable(link))
```

```python
import json
import csv
import pickle
```

```python
with open("offense_statsMinnesota.json", "w") as json_file:
    json.dump(offenseStatsMin, json_file, indent=4)
with open("offense_statsKansas.json", "w") as json_file:
    json.dump(offenseStatsKan, json_file, indent=4)
```

```python
with open('offense_statsMinnesota.pkl', 'wb') as pickle_file:
    pickle.dump(offenseStatsMin, pickle_file)
with open('offense_statsKansas.pkl', 'wb') as pickle_file:
    pickle.dump(offenseStatsKan, pickle_file)
```

```python
with open('offense_stats_nestedMinnesota.csv', 'w', newline='') as csv_file:
    fieldnames = ['url', 'stats']  # Headers for the CSV file
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    writer.writeheader()
    
    for game in offenseStatsMin:
        # Convert the 'stats' (which is a list) into a JSON string
        game['stats'] = json.dumps(game['stats'])
        writer.writerow(game)
with open('offense_stats_nestedKansas.csv', 'w', newline='') as csv_file:
    fieldnames = ['url', 'stats']  # Headers for the CSV file
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    writer.writeheader()
    
    for game in offenseStatsKan:
        # Convert the 'stats' (which is a list) into a JSON string
        game['stats'] = json.dumps(game['stats'])
        writer.writerow(game)
```
